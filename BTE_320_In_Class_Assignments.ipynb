```python
sales = float(input("Enter a sales value: "))
COGS = sales * .40
GrossProfit = sales - COGS

print(f"Sales: {sales}$")
print(f"Cost of Goods Sold: {COGS}$")
print(f"Gross Profit: {GrossProfit}$")


TaxRate = float(input("Enter a tax rate (%): "))
Taxes = GrossProfit * (TaxRate/100)

print(f"Taxes: {Taxes}$")

NetIncome = GrossProfit - Taxes

print(f"Net Income: {NetIncome}$")
```

```python
n1 = float(input("Enter a number: "))
operator = input("Enter an operator (+, -, *, /, **, %, //): ")
n2 = float(input("Enter another number: "))


if operator == "+":
  res = (n1 + n2)
elif operator == "-":
  res = (n1-n2)
elif operator == "*":
  res = (n1*n2)
elif operator == "/":
  res = (n1/n2)
elif operator == "**":
  res = (n1**n2)
elif operator == "%":
  res = (n1%n2)
elif operator == "//":
  res = (n1//n2)
else:
  print("Invalid operation")
  res = None

print(f"Result: {res}")



```
